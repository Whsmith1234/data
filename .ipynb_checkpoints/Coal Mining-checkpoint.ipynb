# Mining data

```
mining = Import["https://greeninfo-network.github.io/mine-tracker/data/data.csv"];
```

```
Table[n  Split[mining[[n]],","][[9]],{n,2,100,1}]
```

```
coords = Table[
a = Split[mining[[n]],","];
GeoPosition[
{a[[14]][[1]], a[[15]][[1]]
}],
{n,2,Length[mining],1}
];
labels = Table[
a = Split[mining[[n]],","];
a[[9]][[1]],
{n,2,Length[mining],1}
];
labeledCoords = MapThread[Labeled, {coords, labels}];
GeoListPlot[labeledCoords,GeoRange->All]
```

```
Histogram[labels]
```
