# Mining data

```
mining = Import["https://greeninfo-network.github.io/mine-tracker/data/data.csv"];
```

```
coords = Table[
a = Split[mining[[n]],","];
GeoPosition[
{a[[14]][[1]], a[[15]][[1]]
}],
{n,2,Length[mining],1}
];
labels = Table[
a = Split[mining[[n]],","];
a[[9]][[1]],
{n,2,Length[mining],1}
];
labeledCoords = MapThread[Labeled, {coords, labels}];
```

## Surface vs Underground Mines

```
GeoListPlot[labeledCoords,GeoRange->All]
```

```
types = Tally[labels];
typesOfMines = Table[types[[n]][[1]], {n,1,Length[types],1}];
types = Table[types[[n]][[2]], {n,1,Length[types],1}];
labeledBars = MapThread[Labeled, {types, typesOfMines}];
```

```
Labeled[PieChart[labeledBars],"Types"]
```

```
surface[x_] := If[labels[[x]] == "surface", coords[[x]]];
underground[x_] := If[labels[[x]] == "underground", coords[[x]]];
surfaceAndUnderground[x_]:= If[labels[[x]] == "underground_and_surface", coords[[x]]];
```

```
n = 1;
surfaceCoords ={};
undergroundCoords ={};
surfaceAndUndergroundCoords = {};
While[n<Length[mining],
AppendTo[surfaceCoords, surface[n]];
AppendTo[undergroundCoords, underground[n]];
AppendTo[surfaceAndUndergroundCoords, surfaceAndUnderground[n]];
n=n+1;
]
```

```
surfaceCoords = DeleteCases[surfaceCoords, Null];
undergroundCoords = DeleteCases[undergroundCoords, Null];
surfaceAndUndergroundCoords = DeleteCases[surfaceAndUndergroundCoords, Null];
```
